Import Libraries..

```
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
url = 'https://raw.githubusercontent.com/anhductn2001/House_price_predicted/main/train.csv'
df = pd.read_csv(url,index_col='Id')
df.head()
```

```
df.info()
```

<h4>Drop các trường có nhiều giá trị bị thiếu hơn</h4>

```
df = df.drop(['PoolQC','Alley','MiscFeature','Fence','FireplaceQu'],axis=1)
```

```
df.head()
```

<h3>Lấy các cột có giá trị bị thiếu, số lượng giá trị thiếu và kiểu giá trị của cột</h3>

```
for i in df.columns:
  if (df[i].isnull().sum()):
    print(i,df[i].isnull().sum(),df[i].dtype)
```


<h3>Điền các ô dữ liệu còn thiếu bằng giá trị xuất hiện thường xuyên nhất trong cột (mode) và trung bình (mean)</h3>

```
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
df['MasVnrArea'].fillna(df['MasVnrArea'].mean(),inplace=True)
df['MasVnrType'].fillna(df['MasVnrType'].mode()[0],inplace=True)
df['BsmtQual'].fillna(df['BsmtQual'].mode()[0],inplace=True)
df['BsmtCond'].fillna(df['BsmtCond'].mode()[0],inplace=True)
df['BsmtExposure'].fillna(df['BsmtExposure'].mode()[0],inplace=True)
df['BsmtFinType1'].fillna(df['BsmtFinType1'].mode()[0],inplace=True)
df['BsmtFinType2'].fillna(df['BsmtFinType2'].mode()[0],inplace=True)
df['Electrical'].fillna(df['Electrical'].mode()[0],inplace=True)
df['GarageType'].fillna(df['GarageType'].mode()[0],inplace=True)
df['GarageYrBlt'].fillna(df['GarageYrBlt'].mean(),inplace=True)
df['GarageFinish'].fillna(df['GarageFinish'].mode()[0],inplace=True)
df['GarageQual'].fillna(df['GarageQual'].mode()[0],inplace=True)
df['GarageCond'].fillna(df['GarageCond'].mode()[0],inplace=True)
```

```
df.shape
```

<h1>Test dataframe</h1>


```

test_df = pd.read_csv('https://raw.githubusercontent.com/anhductn2001/House_price_predicted/main/test.csv')
test_df.head()
```

```
test_df.info()
```

```
test_df = test_df.drop(['PoolQC','Alley','MiscFeature','Fence','FireplaceQu'],axis=1)
```

<h4> Lấy các cột dạng objects </h4>

```
column_list = []
for col in df.columns:
  if df[col].dtype=='O':
    column_list.append(col)
len(column_list)
column_list
```

<h3>Convert sang one hot encoding</h3>

```
def convert_cat_to_dummy(cols):
  df_final = final_df
  i = 0
  for l in cols:
    print(l)
    df1 = pd.get_dummies(final_df[l],drop_first=True)
    final_df.drop([l],axis=1,inplace=True)
    if i == 0:
      df_final=df1.copy()
    else:
      df_final = pd.concat([df_final,df1],axis=1)
    i+=1
  df_final = pd.concat([final_df,df_final],axis=1)
  return df_final
```

```
main_df = df.copy()
```

<h2>Kết hợp tập dữ liệu trước khi gọi hàm one hot encoding </h2>

Vì 1 số trường trong 2 tập không khớp.

```
final_df = pd.concat([df,test_df],axis=0)
final_df.shape
```

```
final_df = convert_cat_to_dummy(column_list)
```

```
final_df.shape
```

<h2>Remove the duplicate columns</h2>

```
final_df = final_df.loc[:,~final_df.columns.duplicated()]
```

```
final_df.shape
```

```
main_df.shape
```

<h2>Seperate the train and test data frame</h2> 

```
df_train = final_df.iloc[:1460,:]
df_test = final_df.iloc[1460:,:]
```

```
df_test.shape
df_test.head()
```

<h1>Sử dụng XGBoost-Regression model</h1>


```
import xgboost
classifer = xgboost.XGBRegressor()
x = df_train.drop(['SalePrice'],axis = 1)
y = df_train['SalePrice']
df_test.drop(['SalePrice'],axis=1,inplace=True)
classifer.fit(x,y)
```

```
y_pred = classifer.predict(df_test)
```

```
y_pred
```

<h1>Export the prediction to csv</h1>

```
output_pred = pd.DataFrame(y_pred)
sub = pd.read_csv('https://raw.githubusercontent.com/anhductn2001/House_price_predicted/main/sample_submission.csv')
dataset = pd.concat([sub['Id'],output_pred],axis=1)
dataset.columns=['Id','SalePrice']
dataset.to_csv('sample_submission.csv',index=False)
```

Accuracy 

```
y_train_pred = classifer.predict(x)
from sklearn.metrics import r2_score
r2_score(y, y_train_pred)
```
